```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.set_ylim([0,10000])
ax1.set_xlim([0,100])
ax1.set_xlabel("x")
ax1.set_ylabel("z")
ax1.plot(x,z)

ax2=fig.add_axes([.2,.5,.4,.4])
ax2.set_ylim([30,50])
ax2.set_xlim([20,22])
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("zoom")
ax2.plot(x,y)
```

```python
fig, pl=plt.subplots(1,2)
```

```python
pl[0].plot(x,y,'b--',lw=3)
pl[1].plot(x,z,'r',lw=4)
```

```python
fig
plt.tight_layout()
```

```python
fig, pl=plt.subplots(1,2,figsize=(13,2))
pl[0].plot(x,y,'b',lw=4)
pl[1].plot(x,z,'r--',lw=3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],kde=False)
```

```python
sns.boxplot(x='class',y='age',data=titanic)
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='coolwarm')
```

```python
sns.countplot(x='sex',data=titanic,palette='coolwarm')
```

```python
sns.heatmap(titanic.corr())
```

```python
sns.barplot(x='age',y='sex',data=titanic)
```

```python
g=sns.FacetGrid(titanic,col='sex')
g.map(plt.hist,'age')
```
